## schema evolution

```sql
%python
data=[(1,'Virat'),(2,'Rohit')]
schema="id int, name string"
df=spark.createDataFrame(data,schema)
df.createOrReplaceTempView("source")
```

```sql
select * from source
```

```sql
create table dev.naval.cricket (id int, name string)
```

```sql
%python
df.write.mode("overwrite").saveAsTable("dev.naval.cricket")
```

```sql
insert into dev.naval.cricket select * from source
```

```sql
select * from dev.naval.cricket 
```

```sql
%python
data=[(3,'Rahul','Batsman'),(4,'hardik','Batsman')]
schema="id int, name string, Role String"
df=spark.createDataFrame(data,schema)
```

```sql
%python
df.write.mode("append").saveAsTable("dev.naval.cricket")
```

```sql
%python
df.write.mode("append").option("mergeSchema",True).saveAsTable("dev.naval.cricket")
```
