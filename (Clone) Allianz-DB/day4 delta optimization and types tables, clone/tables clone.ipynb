```sql
Clone: Copy of your object (table )
```

```sql
1. Deep Clone 
2. Shallow clone 
```

```sql
CREATE OR REPLACE TABLE dev.naval.emp_clone_deep
DEEP CLONE dev.naval.emp;
```

```sql
CREATE OR REPLACE TABLE dev.naval.emp_clone_shallow
SHALLOW CLONE dev.naval.emp;
```

```sql
select * from dev.naval.emp_clone_shallow
```

```sql
delete from dev.naval.emp_clone_shallow where id= 10
```

```sql
desc history dev.naval.emp_clone_shallow
```
